```python
from sympy import Symbol, Derivative, simplify
# Define the symbol x
x = Symbol('x')
# Define the functions
y1 = 3*x**2 - 5*x + 4
y2 = x - x**3
# Calculate the derivatives using SymPy's Derivative function
dy1_dx = Derivative(y1, x).doit()
dy2_dx = Derivative(y2, x).doit()
# Simplify the derivatives
dy1_dx_simplified = simplify(dy1_dx)
dy2_dx_simplified = simplify(dy2_dx)
print("Derivative of y1 = 3x^2 - 5x + 4:", dy1_dx_simplified)
print("Derivative of y2 = x - x^3:", dy2_dx_simplified)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Define the original function and its derivative
def f(x):
    return -4*x + 4
def f_prime(x):
    return -4
# Generate x values
x = np.linspace(-1, 2, 100)
# Calculate y values for f(x) and f'(x)
y = f(x)
y_prime = f_prime(x)
# Plot the graphs
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x)')
plt.plot(x, y_prime, label='f\'(x)', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original Function and its Gradient')
plt.legend()
plt.grid(True)
plt.show()
```
