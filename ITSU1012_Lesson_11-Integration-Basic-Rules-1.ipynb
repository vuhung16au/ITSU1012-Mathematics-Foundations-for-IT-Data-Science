Worked Example 7<br>
Find the equation of the curve g(x),<br>
given that g'(x) = 3*sqrt(x) + 2 and the curve passes through the point (1, 2)

```python
g_prime = 3*sp.sqrt(x) + 2
g = sp.integrate(g_prime, x) + sp.symbols('C')
C_value = sp.solve(g.subs(x, 1) - 2, 'C')[0]
g = g.subs('C', C_value)
print(g)
```

Worked Example 8<br>
If a curve has a stationary point (2, 3) and g'(x) = 2*x - k , <br>
where k is a constant, <br>
find the value of k <br>
find the value of g(x) at x = 1

```python
k = sp.symbols('k')
g_prime = 2*x - k
g = sp.integrate(g_prime, x) + sp.symbols('C')
# Given stationary point (2, 3)
C_value = sp.solve(g.subs(x, 2) - 3, 'C')[0]
g = g.subs('C', C_value)
# g'(2) = 0 for stationary point
k_value = sp.solve(g_prime.subs(x, 2), k)[0]
# Substitute k value back into g(x)
g = g.subs(k, k_value)
# Find g(1)
g_at_1 = g.subs(x, 1)
print(f"k = {k_value}") # should be 4
print(f"g(x) = {g}") # should be x^2 - 4*x + 7
print(f"g(1) = {g_at_1}") # should be 4 
```

Worked Example 20<br>
Draw y = 1/x with x in [1, 4], <br>
Calcuale the area between the curve and the x-axis

```python
import matplotlib.pyplot as plt
# Define the function
def func(x):
    return 1/x
# Define the range
x_vals = np.linspace(1, 4, 400)
y = func(x_vals)
# Plot the function
plt.plot(x_vals, y, 'r', label='y = 1/x')
plt.fill_between(x_vals, y, color='orange', alpha=0.2)
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 1/x')
plt.legend()
plt.grid(True)
plt.show()
# Calculate the area under the curve
f = 1/x
F = sp.integrate(f, (x, 1, 4))
print(F)  # should be ln(4)
```

Worked Example 21<br>
y = x^2 - 4*x<br>
draw the curve y = x^2 - 4*x with x in [0, 4]<br>
Calculate the area between the curve and the x-axis from 1 to 3 <br>
Define the function

```python
def func(x):
    return x**2 - 4*x
# Define the range
x_vals = np.linspace(0, 4, 400)
y = func(x_vals)
# Plot the function
plt.plot(x_vals, y, 'b', label='y = x^2 - 4*x')
plt.fill_between(x_vals, y, where=[(xi >= 1 and xi <= 3) for xi in x_vals], color='orange', alpha=0.2)
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^2 - 4*x')
plt.legend()
plt.grid(True)
plt.show()
# Calculate the area under the curve from 1 to 3
f = x**2 - 4*x
# Take the absolute value of the function
F = abs(sp.integrate(f, (x, 1, 3)))
print(F)  
```

Worked Example 22<br>
y = x^3 - 4*x<br>
Plot the graph of y = x^3 - 4*x with x in [-3, 3]<br>
Calculate the area between the curve and the x-axis from -2 to 2<br>
Define the function

```python
def func(x):
    return x**3 - 4*x
# Define the range
x_vals = np.linspace(-3, 3, 400)
y = func(x_vals)
# Plot the function
plt.plot(x_vals, y, 'g', label='y = x^3 - 4*x')
plt.fill_between(x_vals, y, where=[(xi >= -2 and xi <= 2) for xi in x_vals], color='orange', alpha=0.2)
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^3 - 4*x')
plt.legend()
plt.grid(True)
plt.show()
# Calculate the area under the curve from -2 to 2
f = x**3 - 4*x
# Take the absolute value of the function
F = abs(sp.integrate(f, (x, -2, 0))) + abs(sp.integrate(f, (x, 0, 2)))
print(F)
```
