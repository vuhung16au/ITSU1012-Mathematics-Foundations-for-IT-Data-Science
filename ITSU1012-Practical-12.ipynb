Problem 03

```python
from sympy import *
# Define the variable k
k = Symbol('k', real=True)
integral = integrate(3 + (4 - x)**2, (x, 0, k))
# Set the integral equal to 30 and solve for k
equation = Eq(integral, 30)
solution = solve(equation, k)
real_solution = [sol.evalf() for sol in solution if sol.is_real]
print(real_solution)
```
