*Practical 01*

Vector v has a value of [1, 2] but then a transformation happens.<br>
Vector i lands at [2, 0] and vector j lands at [0, 1.5]<br>
Where does vector land?

```python
v = array([1, 2])
i_hat = array([2, 0])
j_hat = array([0, 1.5])
```

fix the line 

```python
basic = array([i_hat, j_hat])
```

transform vector v into w 

```python
w = np.dot(basic, v)
```

```python
print(w) # [2. 3.]
```

*Practical 03*

A transformation of vector i lands at [1, 0]<br>
and vector j lands at [2, 2].<br>
What is the determinant of the transformation matrix?

```python
i_hat = array([1, 0])
j_hat = array([2, 2])
```

```python
basic = array([i_hat, j_hat]).transpose()
```

```python
determinant = det(basic)
```

```python
print(determinant) # 2.0
```
