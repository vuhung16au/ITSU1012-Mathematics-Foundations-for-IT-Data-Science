```python
def calculate_limit(num_terms):
  """Calculates the sum of the series up to the given number of terms.
```

```python
  Args:
    num_terms: The number of terms to include in the sum.
```

```python
  Returns:
    The sum of the series.
  """
  sum = 0
  for i in range(1, num_terms + 1):
    sum += 1 / (2 ** i)
  return sum
# Set the desired number of terms
num_terms_list = [1, 2, 8, 10, 20, 50, 75, 100, 1000, 10000, 50000]
for num_terms in num_terms_list:
  limit_value = calculate_limit(num_terms)
  print(f"Sum of the series up to {num_terms} terms: {limit_value}")
```

```python
def calculate_limit(x_value):
  """Calculates the value of f(x) = x + 3 for the given x value.
  Args:
    x_value: The value of x.
  Returns:
    The value of f(x).
  """
  return x_value + 3
# Set the desired x values
x_values = [1.95, 1.99, 1.995, 1.9995, 2, 2.005, 2.01, 2.05]
for x in x_values:
  limit_value = calculate_limit(x)
  print(f"f({x}) = {limit_value}")
# Define the function
def calculate_limit(x_value):
    return x_value + 3
# Generate x values around 2
x_values = np.linspace(1.95, 2.05, 400)
y_values = calculate_limit(x_values)
# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = x + 3')
plt.scatter([1.995, 2, 2.005], [calculate_limit(1.995), calculate_limit(2), calculate_limit(2.005)], color='red')
# Annotate the points
for x in [1.995, 2, 2.005]:
    plt.annotate(f'({x}, {calculate_limit(x)})', (x, calculate_limit(x)), textcoords="offset points", xytext=(0,10), ha='center')
plt.title('Graph of f(x) = x + 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def f(x):
  """Defines the function f(x) = (x^2 - x) / (x - 1).
  Args:
    x: The input value.
  Returns:
    The output value of f(x).
  """
  if x != 1:
    return (x**2 - x) / (x - 1)
  else:
    return None  # Handle the discontinuity at x = 1
# Using sympy to find the limit as x approaches 1
x = sp.symbols('x')
f = (x**2 - x) / (x - 1)
# Calculate the limit as x approaches 1
limit_value = sp.limit(f, x, 1)
print(f"The limit of f(x) as x approaches 1 is: {limit_value}")
# Define the range of x values around 1
x_values = np.linspace(0.9, 1.1, 100)
# Calculate the corresponding y values
y_values = [f(x) for x in x_values]
# Plot the graph
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = (x^2 - x) / (x - 1)')
plt.grid(True)
plt.show()
# Find the limit as x approaches 1
limit_value = 1  # The limit is 1, as calculated analytically
print("The limit of f(x) as x approaches 1 is:", limit_value)
```

ITSU1012 Lesson 9 - Worked Example 06

```python
from sympy import Symbol, limit, Piecewise
# Define the symbol x
x = Symbol('x')
# Define the hybrid function f(x)
f = Piecewise((x**2, x <= 2), (x - 1, x > 2))
# Calculate the left-hand limit
left_limit = limit(f, x, 2, dir='-')
# Calculate the right-hand limit
right_limit = limit(f, x, 2, dir='+')
print("Left-hand limit:", left_limit)
print("Right-hand limit:", right_limit)
# Check if the limit exists
if left_limit == right_limit:
  print("The limit exists and is equal to:", left_limit)
else:
  print("The limit does not exist.")
```

ITSU1012 Lesson 9 - Worked Example 16

```python
from sympy import Symbol, Derivative, simplify
# Define the symbol x and u
x, u = Symbol('x'), Symbol('u')
# Define y as a function of u
y = u**3
# Define u as a function of x
u = 3*x - 2
# Calculate dy/du
dy_du = Derivative(y, u).doit()
# Calculate du/dx
du_dx = Derivative(u, x).doit()
# Calculate dy/dx using the chain rule
dy_dx = dy_du * du_dx
# Substitute u back in terms of x
dy_dx = dy_dx.subs(u, 3*x - 2)
# Simplify the expression
dy_dx = simplify(dy_dx)
print("dy/du:", dy_du)
print("du/dx:", du_dx)
print("dy/dx:", dy_dx)
```

ITSU1012 Lesson 9 - Worked Example 26

```python
from sympy import Symbol, Derivative, simplify
# Define the symbol x
x = Symbol('x')
# Define u and v as functions of x
u = 3*x - 1
v = x**2 + 4*x + 3
# Calculate du/dx and dv/dx
du_dx = Derivative(u, x).doit()
dv_dx = Derivative(v, x).doit()
# Calculate dy/dx using the product rule
dy_dx = u * dv_dx + v * du_dx
# Substitute u and v back in terms of x
dy_dx = dy_dx.subs([(u, 3*x - 1), (v, x**2 + 4*x + 3)])
# Simplify the expression
dy_dx = simplify(dy_dx)
print("du/dx:", du_dx)
print("dv/dx:", dv_dx)
print("dy/dx:", dy_dx)
```

ITSU1012 Lesson 9 - Worked Example 27

```python
from sympy import Symbol, Derivative, simplify
# Define the symbol x
x = Symbol('x')
# Define u and v as functions of x
u = 3 - x
v = x**2 + 4*x
# Calculate du/dx and dv/dx
du_dx = Derivative(u, x).doit()
dv_dx = Derivative(v, x).doit()
# Calculate dy/dx using the quotient rule
dy_dx = (v * du_dx - u * dv_dx) / v**2
# Substitute u and v back in terms of x
dy_dx = dy_dx.subs([(u, 3 - x), (v, x**2 + 4*x)])
# Simplify the expression
dy_dx = simplify(dy_dx)
print("du/dx:", du_dx)
print("dv/dx:", dv_dx)
print("dy/dx:", dy_dx)
```
