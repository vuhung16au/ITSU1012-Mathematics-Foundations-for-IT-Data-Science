```python
import numpy as np
import sympy as sp
```

Practical 01 <br>
find the following indefinite integral<br>
¥int (x - 2)**2*(x + 3) dx<br>
¥int (4 - 3*x)**(-3) dx 

```python
x = sp.symbols('x')
f = (x - 2)**2*(x + 3)
g = (4 - 3*x)**(-3)
F = sp.integrate(f, x)
G = sp.integrate(g, x)
print(F)
print(G)
```

Practical 03<br>
A curve has the gradient g'(x) = k*sqrt(x) - 3*x**3, <br>
where k is a constant and a stationary point occurs (1, -2)<br>
Find <br>
a) the value of k<br>
b) the equation of the curve<br>
c) g(4)

```python
k = sp.symbols('k')
g = k*sp.sqrt(x) - 3*x**3
# Find the value of k
k = sp.solve(g.subs(x, 1) + 2, k)[0]
g = g.subs(k, k)
# Find the equation of the curve
g = sp.integrate(g, x) + sp.symbols('C')
C_value = sp.solve(g.subs(x, 1) + 2, 'C')[0]
g = g.subs('C', C_value)
# Find g(4)
g_4 = g.subs(x, 4)
print(k)
print(g)
print(g_4)
```

Practical 07<br>
A curve has the gradient dy/dx f'(x) = k*sin(3*x) - 3,<br>
where k is a constant and a stationary point occurs at (pi/2, -2)<br>
Find the value of k<br>
Find the equation of the curve

```python
k = sp.symbols('k')
f_prime = k*sp.sin(3*x) - 3
# Find the value of k
k_value = sp.solve(f_prime.subs(x, sp.pi/2), k)[0]
f_prime = f_prime.subs(k, k_value)
# Find the equation of the curve
f = sp.integrate(f_prime, x) + sp.symbols('C')
C_value = sp.solve(f.subs(x, sp.pi/2) + 2, 'C')[0]
f = f.subs('C', C_value)
print(k_value)
print(f)
# TODO check with mathgpt
```

Practical 14<br>
y = 1 - x^2 <br>
sketch the graphs of y, show all the intercepts. <br>
Find the area enclosed by the curve, the x-axis and the lines from<br>
a) x = 0 to x = 1 <br>
b) x = 1 to x = 2<br>
c) x = 0 to x = 2

```python
import matplotlib.pyplot as plt
# Define the function
y = 1 - x**2
# Sketch the graph
x_vals = np.linspace(-2, 2, 400)
y_vals = 1 - x_vals**2
plt.plot(x_vals, y_vals, label='y = 1 - x^2')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
# Find the area enclosed by the curve, the x-axis and the lines
# a) x = 0 to x = 1
area_a = sp.integrate(y, (x, 0, 1))
print(f"Area from x=0 to x=1: {area_a}") # should be 2/3
# b) x = 1 to x = 2
area_b = sp.integrate(y, (x, 1, 2))
area_b = abs(area_b)
print(f"Area from x=1 to x=2: {area_b}") # should be 4/3
# c) x = 0 to x = 2
area_c = abs(sp.integrate(y, (x, 0, 1))) + abs(sp.integrate(y, (x, 1, 2)))
print(f"Area from x=0 to x=2: {area_c}") # should be 2
```

Practical 15<br>
sketch the graph of y = 1 + 2*sin(2*x) with x in [0, pi]<br>
show al intercepts.<br>
Find the area enclosed by the curve, the x-axis and the lines from<br>
a) x = 0 to x = 7*pi/12<br>
b) x = 7*pi/12 to x = 11*pi/12<br>
c) x = 0 to x = 11*pi/12<br>
Define the function

```python
y = 1 + 2*sp.sin(2*x)
# Sketch the graph
x_vals = np.linspace(0, np.pi, 400)
y_vals = 1 + 2*np.sin(2*x_vals)
plt.plot(x_vals, y_vals, label='y = 1 + 2*sin(2*x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
# Find the area enclosed by the curve, the x-axis and the lines
# a) x = 0 to x = 7*pi/12
area_a = sp.integrate(y, (x, 0, 7*sp.pi/12))
print(f"Area from x=0 to x=7*pi/12: {area_a}")
# b) x = 7*pi/12 to x = 11*pi/12
area_b = abs(sp.integrate(y, (x, 7*sp.pi/12, 11*sp.pi/12)))
print(f"Area from x=7*pi/12 to x=11*pi/12: {area_b}")
# c) x = 0 to x = 11*pi/12
# not correct 
# area_c = sp.integrate(y, (x, 0, 11*sp.pi/12))
# correct answer, should take absolute value of the area
area_c = area_a + area_b
print(f"Area from x=0 to x=11*pi/12: {area_c}")
```
