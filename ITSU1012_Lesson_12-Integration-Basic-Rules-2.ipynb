Worked Example 01 <br>
a) Sketch the graph of the function g(x) = (3 - x)*(2 + x)<br>
b) Find the area bound by the a-xis and the graph of the function 

```python
x = sp.symbols('x')
g = (3 - x) * (2 + x)
# a) Sketch the graph of the function g(x)
x_vals = np.linspace(-4, 4, 400)
g_vals = [g.subs(x, val) for val in x_vals]
plt.plot(x_vals, g_vals, label='g(x) = (3 - x)*(2 + x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Graph of g(x) = (3 - x)*(2 + x)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.show()
g_abs = sp.Abs(g)
# b) Find the area bound by the x-axis and the graph of the function
area = sp.integrate(g, (x, -2, 3))
area_g_abs = sp.integrate(g_abs, (x, -2, 3))
print(f'The area bound by the x-axis and the graph of the function is: {area}')
print(f'The area bound by the x-axis and the graph of the function is: {area_g_abs}')
```

Worked Example 02<br>
a) Find the x-intercepts of y = sin (2*x) over the domain [0, 2*\pi]<br>
b) Calculate the area between the curve, the x-axis and x = 0 and x = \pi<br>
a) Find the x-intercepts of y = sin(2*x) over the domain [0, 2*pi]

```python
x_intercepts = sp.solveset(sp.sin(2 * x), x, domain=sp.Interval(0, 2 * sp.pi))
print(f'The x-intercepts of y = sin(2*x) over the domain [0, 2*pi] are: {x_intercepts}')
# b) Calculate the area between the curve, the x-axis and x = 0 and x = pi
y = sp.sin(2 * x)
area_between_curve_and_x_axis = sp.integrate(sp.Abs(y), (x, 0, sp.pi))
print(f'The area between the curve y = sin(2*x), the x-axis, x = 0 and x = pi is: {area_between_curve_and_x_axis}')
```

Worked Example 03<br>
a) Differentiate log_e(x^2 - 1)<br>
b) Hence, find an antiderivative of x/(x^2 - 1)<br>
c) Find the area between the graph of x/(x^2 - 1), <br>
   the x-axis, x = 2 and x = 3, giving your answer correct to 2 decimal places. <br>
a) Differentiate log_e(x^2 - 1)

```python
f = sp.log(x**2 - 1)
f_prime = sp.diff(f, x)
print(f'The derivative of log_e(x^2 - 1) is: {f_prime}')
# b) Hence, find an antiderivative of x/(x^2 - 1)
integrand = x / (x**2 - 1)
antiderivative = sp.integrate(integrand, x)
print(f'An antiderivative of x/(x^2 - 1) is: {antiderivative}')
# c) Find the area between the graph of x/(x^2 - 1), 
#    the x-axis, x = 2 and x = 3, giving your answer correct to 2 decimal places.
area = sp.integrate(sp.Abs(integrand), (x, 2, 3))
area_rounded = round(area.evalf(), 2)
print(f'The area between the graph of x/(x^2 - 1), the x-axis, x = 2 and x = 3 is: {area_rounded}') # should be 0.49
```

Worked Example 04<br>
Write a python program to <br>
i) Illustrate the worked example 4 <br>
ii) Draw the graph of y = 2x + 1 and y = x on the same diagram <br>
iii) Indicate the area between y = 2x + 1, y = x, x = 0, x = 2. Calculate this area. 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
# Define the functions f(x) = 2x + 1 and g(x) = x
def f(x):
    return 2*x + 1
def g(x):
    return x
# Create a range of x values
x = np.linspace(0, 2, 100)
# Calculate the corresponding y values for each function
y1 = f(x)
y2 = g(x)
# Plot the functions
plt.plot(x, y1, label='y = 2x + 1')
plt.plot(x, y2, label='y = x')
# Fill the area between the curves
plt.fill_between(x, y1, y2, color='lightblue', alpha=0.5)
# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Worked Example 4')
# Set the limits of the x and y axes
plt.xlim(0, 2)
plt.ylim(0, 5)
# Add a legend
plt.legend()
# Calculate the area using numerical integration
area, _ = quad(lambda x: f(x) - g(x), 0, 2)
print("Area between the curves:", area) # should be 4 
# Show the plot
plt.show()
```

Worked Example 05<br>
a) Find the values of x where the function y = x and y = x^2 -2 intersect<br>
b) Sketch the graphs on the same axes (using Python mathplot)<br>
c) Hence, find the area bound by the curves<br>
a) Find the values of x where the function y = x and y = x^2 - 2 intersect

```python
x = sp.symbols('x')
intersection_points = sp.solve(x - (x**2 - 2), x)
print(f'The values of x where the functions y = x and y = x^2 - 2 intersect are: {intersection_points}')
# b) Sketch the graphs on the same axes (using Python matplotlib)
x_vals = np.linspace(-2, 3, 400)
y1_vals = x_vals
y2_vals = x_vals**2 - 2
plt.plot(x_vals, y1_vals, label='y = x')
plt.plot(x_vals, y2_vals, label='y = x^2 - 2')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Graphs of y = x and y = x^2 - 2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
# c) Hence, find the area bound by the curves
area_between_curves = sp.integrate(sp.Abs(x - (x**2 - 2)), (x, intersection_points[0], intersection_points[1]))
print(f'The area bound by the curves y = x and y = x^2 - 2 is: {area_between_curves.evalf()}') # should be 4.5
```

Worked Example 07<br>
if f'(x) = 4*x*e^(x^2) and f(0) = 5, find f(x)<br>
if f'(x) = 4*x*e^(x^2) and f(0) = 5, find f(x)

```python
f_prime = 4 * x * sp.exp(x**2)
f = sp.integrate(f_prime, x) + sp.symbols('C')
C_value = sp.solve(f.subs(x, 0) - 5, 'C')[0]
f = f.subs('C', C_value)
print(f'The function f(x) is: {f}')
```
