Practical 01 


<br>
using sympy<br>
If f(x) = x^2 +2*x -8 find:<br>
d) the average rate of change between x = 3 and x = 6.<br>
e) a function which describes the rate of change.<br>
f) the instantaneous rate of change at x = 3.<br>
<br>
Define the variable and the function

```python
x = sp.symbols('x')
f = x**2 + 2*x - 8
# d) the average rate of change between x = 3 and x = 6
x1, x2 = 3, 6
average_rate_of_change = (f.subs(x, x2) - f.subs(x, x1)) / (x2 - x1)
print(f"Average rate of change between x = {x1} and x = {x2}: {average_rate_of_change}")
# e) a function which describes the rate of change
rate_of_change_function = sp.diff(f, x)
print(f"Function which describes the rate of change: {rate_of_change_function}")
# f) the instantaneous rate of change at x = 3
instantaneous_rate_of_change_at_3 = rate_of_change_function.subs(x, 3)
print(f"Instantaneous rate of change at x = 3: {instantaneous_rate_of_change_at_3}")
```

Practical 02


 <br>
Using sympy, <br>
If the position of a vehicle is given by the equation:<br>
x(t) = 2*t^2 - 20*t, find:<br>
d) the average rate of change (velocity)<br>
between t = 1 and t = 3.<br>
e) a function which describes the velocity.<br>
f) the instantaneous velocity at t = 3.<br>
<br>
Define the variable and the function

```python
t = sp.symbols('t')
x_t = 2*t**2 - 20*t
# d) the average rate of change (velocity) between t = 1 and t = 3
t1, t2 = 1, 3
average_velocity = (x_t.subs(t, t2) - x_t.subs(t, t1)) / (t2 - t1)
print(f"Average rate of change (velocity) between t = {t1} and t = {t2}: {average_velocity}")
# e) a function which describes the velocity
velocity_function = sp.diff(x_t, t)
print(f"Function which describes the velocity: {velocity_function}")
# f) the instantaneous velocity at t = 3
instantaneous_velocity_at_3 = velocity_function.subs(t, 3)
print(f"Instantaneous velocity at t = 3: {instantaneous_velocity_at_3}")
```

Sketch both of the functions showing stationary points and y intercepts.<br>

Practical 11


 <br>
A projectile is fired with a trajectory equation of:<br>
H(x) = 20 + 8*x - x^2,<br>
x in [0, 10],<br>
where H is the<br>
vertical height and x is the horizontal displacement<br>
(both distances measured in metres).<br>
What is the maximum height and at what<br>
horizontal displacement does it occur?<br>
Sketch the path of the projectile.<br>
<br>
Define the function

```python
H = 20 + 8*x - x**2
# Find the gradient of the function
gradient_H = sp.diff(H, x)
# Locate the stationary points
zero_gradient_H = sp.solve(gradient_H, x)
stationary_points_H = [(x_val, H.subs(x, x_val)) for x_val in zero_gradient_H if 0 <= x_val <= 10]
# Find the maximum height and the corresponding horizontal displacement
max_height_point = max(stationary_points_H, key=lambda point: point[1])
max_height, horizontal_displacement = max_height_point[1], max_height_point[0]
print(f"Maximum height: {max_height} metres at horizontal displacement: {horizontal_displacement} metres")
# Sketch the path of the projectile
x_vals = np.linspace(0, 10, 400)
H_vals = [H.subs(x, val) for val in x_vals]
plt.figure(figsize=(7, 6))
plt.plot(x_vals, H_vals, label='H(x) = 20 + 8*x - x^2')
plt.scatter(horizontal_displacement, max_height, color='red', zorder=5)
plt.title('Trajectory of the Projectile')
plt.xlabel('Horizontal Displacement (metres)')
plt.ylabel('Vertical Height (metres)')
plt.legend()
plt.grid(True)
plt.show()
```

Practical 12


 <br>
The profit ($p) to be earned by a small company<br>
per day is related to the number of employees (x),<br>
according to the equation<br>
p(x) = 2000 + 216*x -2*x^3, <br>
x in [1, 10].<br>
What is the maximum number of employees for<br>
maximum profit? What is the maximum profit?<br>
Sketch the profit function.<br>
<br>
Define the function

```python
p = 2000 + 216*x - 2*x**3
# Find the gradient of the function
gradient_p = sp.diff(p, x)
# Locate the stationary points
zero_gradient_p = sp.solve(gradient_p, x)
stationary_points_p = [(x_val, p.subs(x, x_val)) for x_val in zero_gradient_p if 1 <= x_val <= 10]
# Find the maximum profit and the corresponding number of employees
max_profit_point = max(stationary_points_p, key=lambda point: point[1])
max_profit, max_employees = max_profit_point[1], max_profit_point[0]
print(f"Maximum profit: ${max_profit} with {max_employees} employees")
# Sketch the profit function
x_vals = np.linspace(1, 10, 400)
p_vals = [p.subs(x, val) for val in x_vals]
plt.figure(figsize=(7, 6))
plt.plot(x_vals, p_vals, label='p(x) = 2000 + 216*x - 2*x^3')
plt.scatter(max_employees, max_profit, color='red', zorder=5)
plt.title('Profit Function')
plt.xlabel('Number of Employees')
plt.ylabel('Profit ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import sympy as sp
# Define the variables
x = sp.Symbol('x')
# Define the area function
area = x * (480 - 2*x)
# Find the derivative of the area function
derivative = area.diff(x)
# Find the critical point by setting the derivative to zero
critical_point = sp.solve(derivative, x)[0]
# Calculate the width and length at the critical point
width = critical_point
length = 480 - 2*width
# Calculate the maximum area
max_area = area.subs(x, critical_point)
print(f"Width: {width} meters")
print(f"Length: {length} meters")
print(f"Maximum Area: {max_area} square meters")
```

```python
import sympy as sp
# Define the variable
n = sp.Symbol('n', real=True)
# Define the function T(n)
T = n**2 + 1/n**2
# Find the derivative of T(n)
dT = T.diff(n)
# Find the critical points by setting the derivative to zero
critical_points = sp.solve(dT, n)
# Filter real critical points
real_critical_points = [point.evalf() for point in critical_points if point.is_real]
# Check if the critical points give a minimum or maximum
d2T = dT.diff(n)
for point in real_critical_points:
    second_derivative = d2T.subs(n, point)
    if second_derivative > 0:
        print(f"The number {point} gives a minimum value for T(n)")
    else:
        print(f"The number {point} does not give a minimum value for T(n)")
```

```python
from sympy import Symbol, integrate, solve, Eq
# Define the derivative of the function
f_prime = 4*x**3 - 4*x
# Integrate the derivative to get the general form of the function
x = Symbol('x')
f = integrate(f_prime, x)
# Add the constant of integration
C = Symbol('C')
f += Symbol('C')
# Use the initial condition f(2) = 15 to solve for C
C_value = solve(Eq(f.subs(x, 2), 15), C)[0]
# Substitute C with its value in the function
f = f.subs(C, C_value)
print(f"The rule for the function f(x) is: {f}")
```
