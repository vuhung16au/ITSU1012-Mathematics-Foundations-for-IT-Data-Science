# Practical 06

```python
starting_point = np.array([0, 0])
```

move starting_point 30 to the right, at an angle of elevation of 30 degree

Define the angle and distance

```python
angle_degrees = 30
distance = 30
```

Convert angle to radians

```python
angle_radians = np.radians(angle_degrees)
```

Calculate the displacement vector

```python
displacement_vector = np.array([distance * np.cos(angle_radians), distance * np.sin(angle_radians)])
```

Calculate the new position

```python
new_position = starting_point + displacement_vector
```

```python
print(f"The new position of the point after moving 30 units to the right at an angle of 30 degrees is {new_position}")
```

and then move horizontally to the right 400.

Move horizontally to the right 400 units

```python
final_position = new_position + np.array([400, 0])
```

```python
print(f"The final position of the point after moving horizontally to the right 400 units is {final_position}")
```

Plot the starting point

```python

# Plot the starting point
plt.plot(*starting_point, 'ro', label='Starting Point (0,0)')

# Plot the displacement vector
plt.quiver(*starting_point, *displacement_vector, angles='xy', scale_units='xy', scale=1, color='b', label='Displacement Vector (30 units at 30 degrees)')

# Plot the new position
plt.plot(*new_position, 'go', label=f'New Position {new_position}')

# Plot the horizontal movement vector
horizontal_vector = np.array([400, 0])
plt.quiver(*new_position, *horizontal_vector, angles='xy', scale_units='xy', scale=1, color='m', label='Horizontal Movement (400 units)')

# Plot the final position
plt.plot(*final_position, 'co', label=f'Final Position {final_position}')

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Set the limits of the plot
plt.xlim(-50, 450)
plt.ylim(-50, 100)

# Add grid, legend, and labels
plt.grid()

# TODO: Turn legends ON/OFF
# plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Diagram: Displacement and Horizontal Movement')

# Show the plot
plt.show()
```

# Practical 20

A speedboat travel 54km/h in a direction S27°E.<br>
How far<br>
a) South<br>
b) East<br>
of its starting point has it travelled?<br>
Show it on a diagram.

Define the speed and direction

```python
speed = 54  # km/h
angle_degrees = 27  # S27°E
```

Convert angle to radians

```python
angle_radians = np.radians(angle_degrees)
```

Calculate the displacement components

```python
displacement_south = speed * np.cos(angle_radians)
displacement_east = speed * np.sin(angle_radians)
```

```python
print(f"The speedboat has travelled {displacement_south:.2f} km south and {displacement_east:.2f} km east.")
```

Plot the displacement vector

```python
starting_point = np.array([0, 0])
displacement_vector = np.array([displacement_east, -displacement_south])
```

```python
plt.quiver(*starting_point, *displacement_vector, angles='xy', scale_units='xy', scale=1, color='b', label='Displacement Vector (54 km/h at S27°E)')
```

Plot the starting point

```python
plt.plot(*starting_point, 'ro', label='Starting Point (0,0)')
```

Plot the displacement components

```python

# Plot the displacement vector
starting_point = np.array([0, 0])
displacement_vector = np.array([displacement_east, -displacement_south])

plt.quiver(*starting_point, *displacement_vector, angles='xy', scale_units='xy', scale=1, color='b', label='Displacement Vector (54 km/h at S27°E)')

# Plot the starting point
plt.plot(*starting_point, 'ro', label='Starting Point (0,0)')

# Plot the displacement components
plt.quiver(*starting_point, displacement_east, 0, angles='xy', scale_units='xy', scale=1, color='g', label='East Component')
plt.quiver(*starting_point, 0, -displacement_south, angles='xy', scale_units='xy', scale=1, color='r', label='South Component')

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Set the limits of the plot
plt.xlim(-10, 60)
plt.ylim(-60, 10)

# Add grid, legend, and labels
plt.grid()

# TODO: Turn legend ON/OFF
# plt.legend()

plt.xlabel('X-axis (East)')
plt.ylabel('Y-axis (South)')
plt.title('Speedboat Displacement: S27°E at 54 km/h')

# Show the plot
plt.show()
```
