# Problem 10

```python
# Practical 10

# An object is thrown from the top of a high raise building with and equation of trajectory
# H = 8*x - 4*x**2 + 5, x in [0, 4]
# where H is the vertical height above the building and x is the horizontal distance from the building
# (both distances measured in meters).

# Sketch the path and calculate the maximum height and minimum height of the object.

# Define the function H(x)
def H(x):
    return 8*x - 4*x**2 + 5

# Define the x values for the graph
x = np.linspace(0, 4, 400)

# Calculate the y values for the graph
y = H(x)

# Find y_max
# y_max = max(y)

# Create the figure and the axes
fig, ax = plt.subplots()

# Plot the graph
ax.plot(x, y, label='H = 8*x - 4*x**2 + 5')

# Set the x and y axis labels
ax.set_xlabel('x (horizontal distance in meters)')
ax.set_ylabel('H (vertical height in meters)')

# Set the x and y axis limits
ax.set_xlim(0, 4)
ax.set_ylim(0, max(y) + 1)

# Show grid lines on the graph
ax.grid(True)

# Make the x and y axis equal
ax.axis('equal')

# Set the title of the graph
ax.set_title('Trajectory of the Object')

# Add a legend to the graph
ax.legend()

# Make x and y axis line in BOLD and BLACK
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)

# Display the graph
plt.show()

# Calculate the maximum height
x_max = -8 / (2 * -4)
H_max = H(x_max)
print(f"The maximum height of the object is: {H_max} meters")

# Calculate the minimum height (at the endpoints)
H_min = min(H(0), H(4))
print(f"The minimum height of the object is: {H_min} meters")

# End of Practical 10
```

```python
# Practical 11

# Solve the following quadratic equations over C (complex numbers)

print("Solving the quadratic equations over C (complex numbers) using discriminant")

# (a) x**2 + 2*x + 5 = 0
a, b, c = 1, 2, 5
discriminant = cmath.sqrt(b**2 - 4*a*c)
root1 = (-b + discriminant) / (2*a)
root2 = (-b - discriminant) / (2*a)
print(f"The roots of the equation x**2 + 2*x + 5 = 0 are: {root1} and {root2}")

# (b) x**2 - 8*x + 25 = 0
a, b, c = 1, -8, 25
discriminant = cmath.sqrt(b**2 - 4*a*c)
root1 = (-b + discriminant) / (2*a)
root2 = (-b - discriminant) / (2*a)
print(f"The roots of the equation x**2 - 8*x + 25 = 0 are: {root1} and {root2}")

# (c) 4*x**2 - 12*x + 13 = 0
a, b, c = 4, -12, 13
discriminant = cmath.sqrt(b**2 - 4*a*c)
root1 = (-b + discriminant) / (2*a)
root2 = (-b - discriminant) / (2*a)
print(f"The roots of the equation 4*x**2 - 12*x + 13 = 0 are: {root1} and {root2}")

# (d) 4*x**2 - cmath.sqrt(32)*x + 4 = 0
a, b, c = 4, cmath.sqrt(32), 4
discriminant = cmath.sqrt(b**2 - 4*a*c)
root1 = (-b + discriminant) / (2*a)
root2 = (-b - discriminant) / (2*a)
print(f"The roots of the equation 4*x**2 - sqrt(32)*x + 4 = 0 are: {root1} and {root2}")

# Use sympy to solve the quadratic equations over C (complex numbers)

print("Using sympy to solve the quadratic equations over C (complex numbers)")

# (a) x**2 + 2*x + 5 = 0
x = sp.symbols('x')
eq_a = x**2 + 2*x + 5
roots_a = sp.solve(eq_a, x)
print(f"The roots of the equation x**2 + 2*x + 5 = 0 are: {roots_a}")

# (b) x**2 - 8*x + 25 = 0
eq_b = x**2 - 8*x + 25
roots_b = sp.solve(eq_b, x)
print(f"The roots of the equation x**2 - 8*x + 25 = 0 are: {roots_b}")

# (c) 4*x**2 - 12*x + 13 = 0
eq_c = 4*x**2 - 12*x + 13
roots_c = sp.solve(eq_c, x)
print(f"The roots of the equation 4*x**2 - 12*x + 13 = 0 are: {roots_c}")

# (d) 4*x**2 - sp.sqrt(32)*x + 4 = 0
eq_d = 4*x**2 - sp.sqrt(32)*x + 4
roots_d = sp.solve(eq_d, x)
print(f"The roots of the equation 4*x**2 - sqrt(32)*x + 4 = 0 are: {roots_d}")

# End of Practical 11
```
