*Example 09.04 Matrix LU Decomposition*

LU Decomposition is a matrix decomposition technique that factors a matrix into <br>
a lower triangular matrix and an upper triangular matrix.

Declare a matrix

```python
A = array([[4, 2, 4],
            [5, 3, 7],
            [9, 3, 6]])
```

Perform LU decomposition with scipy.linalg.lu

```python
P, L, U = lu(A)
```

```python
print(P) # prints [[0. 1. 0.] [0. 0. 1.] [1. 0. 0.]]
print(L) # prints [[1. 0. 0.] [0.44444444 1. 0.] [0.55555556 0.5 1.]]
print(U) # prints [[9. 3. 6.] [0. 0.33333333 1.66666667] [0. 0. 1.]]
print(P @ L @ U) # prints [[4. 2. 4.] [5. 3. 7.] [9. 3. 6.]]
print(np.allclose(A, P @ L @ U)) # prints True
```
