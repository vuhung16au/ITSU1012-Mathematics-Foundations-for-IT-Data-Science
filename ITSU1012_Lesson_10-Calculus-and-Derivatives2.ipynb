Worked Example 01 <br>
if f(x) = x^3 -6*x^2 -15*x <br>
i) find the value of f'(x)<br>
a) find the values x where the gradient of the curve is zero<br>
b) the stationary points of the curve<br>
Define the variable and function

```python
x = sp.symbols('x')
f = x**3 - 6*x**2 - 15*x
# i) Find the value of f'(x)
f_prime = sp.diff(f, x)
print(f"f'(x) = {f_prime}")
# a) Find the values of x where the gradient of the curve is zero
critical_points = sp.solve(f_prime, x)
print(f"Values of x where the gradient is zero: {critical_points}")
# b) Find the stationary points of the curve
stationary_points = [(point, f.subs(x, point)) for point in critical_points]
print(f"Stationary points: {stationary_points}")
```

Worked Example 02<br>
if y = f(x) = 1/3*x^3 +2*x^2 + 3*x -2<br>
i) find the value of f'(x)<br>
a) find the the stationary points of the curve<br>
b) and determine their nature<br>
Define the function

```python
f2 = (1/3)*x**3 + 2*x**2 + 3*x - 2
# i) Find the value of f'(x)
f2_prime = sp.diff(f2, x)
print(f"f'(x) = {f2_prime}")
# a) Find the stationary points of the curve
critical_points2 = sp.solve(f2_prime, x)
print(f"Values of x where the gradient is zero: {critical_points2}")
# Find the coordinates (x, y) of the stationary points
stationary_points2 = [(point, f2.subs(x, point)) for point in critical_points2]
print(f"Stationary points: {stationary_points2}")
# b) Determine the nature of the stationary points
second_derivative = sp.diff(f2_prime, x)
nature_of_points = [(point, second_derivative.subs(x, point)) for point in critical_points2]
for point, nature in nature_of_points:
    if nature > 0:
        nature_str = "local minimum"
    elif nature < 0:
        nature_str = "local maximum"
    else:
        nature_str = "inconclusive"
    print(f"Stationary point at x = {point} is a {nature_str}")
```

Worked Example 03<br>
f(x) = 5 +4*x -x**2<br>
Sketch the curve of the function<br>
and labelling all intercepts of stationary points

```python
import matplotlib.pyplot as plt
# Define the function
f3 = 5 + 4*x - x**2
# Find the first derivative
f3_prime = sp.diff(f3, x)
# Find the critical points
critical_points3 = sp.solve(f3_prime, x)
# Find the coordinates (x, y) of the stationary points
stationary_points3 = [(point, f3.subs(x, point)) for point in critical_points3]
# Find the x and y intercepts
x_intercepts = sp.solve(f3, x)
y_intercept = f3.subs(x, 0)
# Convert sympy expressions to numpy functions for plotting
f3_np = sp.lambdify(x, f3, 'numpy')
# Generate x values
x_vals = np.linspace(-10, 10, 400)
y_vals = f3_np(x_vals)
# Plot the function
plt.plot(x_vals, y_vals, label='f(x) = 5 + 4x - x^2')
# Plot the stationary points
for point, value in stationary_points3:
    plt.plot(point, value, 'ro')  # red dot
    plt.text(point, value, f'({point}, {value})')
# Plot the x intercepts
for intercept in x_intercepts:
    plt.plot(intercept, 0, 'go')  # green dot
    plt.text(intercept, 0, f'({intercept}, 0)')
# Plot the y intercept
plt.plot(0, y_intercept, 'bo')  # blue dot
plt.text(0, y_intercept, f'(0, {y_intercept})')
# Add labels and legend
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Graph of f(x) = 5 + 4x - x^2')
plt.show()
```

Worked Example 04<br>
A baseball fielder throws the ball so that the equation of the path of the ball is given by<br>
y = f(x) = 1.5 + x - 0.02*x^2<br>
where x (metres) is the horizontal distance travelled by the ball and y (metres) is the <br>
vertical height of the ball above the ground.<br>
a) Find the value of x for which the maximum height is reached (verify that it is a maximum)<br>
b) Find the maximum height reached <br>
Define the function

```python
f4 = 1.5 + x - 0.02*x**2
# Find the first derivative
f4_prime = sp.diff(f4, x)
# a) Find the value of x for which the maximum height is reached
critical_points4 = sp.solve(f4_prime, x)
print(f"Values of x where the gradient is zero: {critical_points4}")
# Verify that it is a maximum by checking the second derivative
second_derivative4 = sp.diff(f4_prime, x)
nature_of_points4 = [(point, second_derivative4.subs(x, point)) for point in critical_points4]
for point, nature in nature_of_points4:
    if nature < 0:
        nature_str = "local maximum"
    elif nature > 0:
        nature_str = "local minimum"
    else:
        nature_str = "inconclusive"
    print(f"Stationary point at x = {point} is a {nature_str}")
# b) Find the maximum height reached
max_height = f4.subs(x, critical_points4[0])
print(f"The maximum height reached is: {max_height}")
```

Worked Example 05

```python
import numpy as np
def max_area(perimeter):
    """
    Calculates the maximum area of a rectangular paddock with the given perimeter.
    Args:
        perimeter: The total length of fencing available.
    Returns:
        The maximum area of the paddock, and the corresponding length and width.
    """
    # Express area in terms of w: A = 240w - 2w^2
    def area_eq(w):
        return 240 * w - 2 * w**2
    # Find the stationary points by taking the derivative and setting it to zero
    def derivative_area_eq(w):
        return 240 - 4 * w
    # Find the stationary point
    stationary_point = derivative_area_eq(0) / 4
    # Calculate the maximum area and corresponding length and width
    max_area = area_eq(stationary_point)
    width = stationary_point
    length = perimeter - 2 * width
    return max_area, length, width
# Example usage:
perimeter = 240
max_area, length, width = max_area(perimeter)
print(f"The maximum area of the paddock is: {max_area} square meters")
print(f"The corresponding length and width are: {length:.2f} meters and {width:.2f} meters")
```

```python
from sympy import Symbol, integrate, solve, Eq
# Define the derivative of the function
f_prime = 3 + 4*x - x**2
# Integrate the derivative to get the general form of the function
x = Symbol('x')
C = Symbol('C')
f = integrate(f_prime, x)
# Add the constant of integration
f += C
# Use the initial condition f(0) = 7 to solve for C
C_value = solve(Eq(f.subs(x, 0), 7), C)[0]
# Substitute C with its value in the function
f = f.subs(C, C_value)
print(f"The rule for the function f(x) is: {f}")
```

Worked Example 07

```python
from sympy import Symbol, integrate, solve, Eq
# Define the variable
t = Symbol('t')
# Define the derivative of the volume function
dVdt = 3*t**2 - 8*t + 1
# Integrate dV/dt to get the general form of the volume function
t = Symbol('t')
V = integrate(dVdt, t)
# Add the constant of integration
V += Symbol('C')
# Use the initial condition V(0) = 6 to solve for C
C_value = solve(Eq(V.subs(t, 0), 6), C)[0]
# Substitute C with its value in the volume function
V = V.subs(C, C_value)
print(f"The volume of the balloon as a function of time is: V(t) = {V}")
# Calculate the volume at t = 1
volume_at_1 = V.subs(t, 1)
print(f"The volume of the balloon at t = 1 is: {volume_at_1} liters")
```
